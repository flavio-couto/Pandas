# CRIANDO COLUNAS NUMERICAS

```python
# valor_por_mes: essa coluna deve conter os gastos mensais de cada imóvel, incluindo aluguel e condomínio;
dados['Valor_por_Mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
#valor_por_ano: essa coluna deve conter os gastos anuais por imóvel, ou seja, IPTU mais 12 meses de aluguel e condomínio.
dados['Valor_por_Ano'] = dados['Valor_por_Mes'] * 12 + dados['IPTU']
dados.head()
```

```python
dados.head()
# Aqui já com as duas colunas que foram criadas
```

# CRIANDO COLUNAS CATEGÓRICAS

```python
# Descrição: essa coluna deve possuir uma sumarização das principais informações: tipo de imóvel, bairro, quantidade de quartos e vagas de garagem;
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' quarto(s) e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem'
dados.head()
# A biblioteca não concatena strings com números, nesse caso as informações que são numéricas tranformamos elas em strings com o 'ASTYPE(STR)' para podermos juntar tudo
```

```python
# Possui_suite: essa deve ser uma coluna que informe apenas se o imóvel possui ou não suítes, sem se importar com a quantidade.
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
pd.read_csv('dados_completos_dev.csv', sep=';')
```

```python

url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
pd.read_csv(url)
notas_atv_4 = pd.read_csv(url)
notas_atv_4.fillna(0, inplace=True)
notas_atv_4
```

```python
# 1) Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles. Crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno.
notas_atv_4['Pontos_Extras'] = notas_atv_4['Notas'] * 0.4
notas_atv_4
# OU notas_atv_4['Pontos_extras'] = notas_atv_4['Notas'].apply(lambda x: x*0.4)
```

```python
# 2) Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.
notas_atv_4['Notas_Finais'] = notas_atv_4['Notas'] + notas_atv_4['Pontos_Extras']
# notas_atv_4['Notas_Finais'] = notas_atv_4['Notas_Finais'].apply(lambda x: 10.00 if x > 10.00 else )
# Com a nota extra alguns alunos passaram de 10.00, estava tentando colocar acima uma condição para que se fosse > 10 ele manter só 10 mas não consegui fazer o else ...
# que era pra retornar o valor da própria célula mas não consegui
notas_atv_4
```

```python
# 3) Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:
notas_atv_4['Aprovado_Final'] = notas_atv_4['Notas_Finais'].apply(lambda x: True if x >= 6 else False)
notas_atv_4
# True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);
# False: caso o aluno esteja reprovado (nota final deve ser menor que 6).
```

```python
# 4) Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.
notas_atv_4.query('Aprovado == False & Aprovado_Final == True ')
```
